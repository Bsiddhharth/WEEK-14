Catplots [Figure Level Function]
1. Categorical Scatterplots
-    Striplot
-    Swarmplot   

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# pass categorical data on x-axis
plt.style.use('fivethirtyeight')
sns.catplot(x='day',y='tip',kind='strip',hue='day',data=tips)
```

```python
#  jitter parameter is used to add a small amount of random noise to the categorical scatter plot points. 
# This helps in better visualization of the distribution of data points when multiple points share the same categorical value.

sns.catplot(x='day',y='tip',hue='day',jitter=0,data=tips)
```

```python
# swarm plot gives you distribution of data
sns.catplot(x='day',y='tip',kind='swarm',hue='day',data=tips)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',hue='sex', data=tips)
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='tip')
```

Categorical distribution plots
- Boxplot
- Violinplot

Box plot- Standardized way of displaying distribution of data based on a five number summary
- minimum, 1st quartile(Q1), median, 3rd quartile(Q3), maximum
- it tells you wether your data is symmetrical or not
- how tightly your data is grouped

```python
 #univariate
```

```python
sns.boxplot(x=tips['tip'])
```

```python
sns.boxplot(x=tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='box',hue='sex',data=tips)
```

Violin plots
- method of plotting numerical data 
- can be considered a combination with of box plot with a kernal density plot (Probability Density Function)
- advantage over boxplot is - it shows entire distribution of data along with 5 figures

```python
sns.violinplot(x=tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='day',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',data=tips)
```

```python
# above the data seems cluttered
# use split parameter to use space efficiently
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='violin',split=True,data=tips)
```

Categorical estimate plots
- Barplot
- Countplot
- Pointplot

```python
#default value of bar plot is to calculate the mean of numercal value
sns.catplot(x='smoker',y='total_bill',kind='bar',hue='smoker',data=tips)
```

```python
# the black bars(error bars) above signifies condifence interval
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',data=tips)
```

```python
# lets change from finding mean to median
# use estimator
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',estimator=np.median,data=tips)
```

Count plots are special case of barplot whre you donot consider any numerical values,
instead count of categorical value 

```python
sns.catplot(x='sex',kind='count',hue='smoker',data=tips)
```

HeatMaps
- graphical representation 
- where individual values contained in a matrix  are represented as colors
- sometime drawn between numerical data but it must be descrete
- usually between categorical data

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
# At 1st build a pivot table or a cross tabulation table
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.1,annot=True,fmt='d')
```

```python
sns.heatmap(x,linewidths=0.5,annot=True,fmt='d',cbar=False,cmap='viridis')
```

```python
sns.heatmap(x,linewidths=0.5,annot=True,fmt='d',cbar=False,cmap='winter')
```

Cluster Maps
- heapmap but heirarchicaly clustered
- on the basis of similarity
  

```python
# clusterinh happenns because of the parameter called metric
# default set to euclidean ie clustered based on the eucledean distance between points
sns.clustermap(x,cbar=False,cmap='summer')
```

```python
sns.clustermap(x,cbar=False,metric='correlation',annot=True,z_score=0)
```

```python
sns.clustermap(x,annot=True,cmap='winter',z_score=0,col_cluster=False)
```

Joint Plot
- used to perform bivariate analysis on numerical data
- multi- panel figure that shows both the bivariate relationship between 2 variables along with univariate distribution of each on seperate axis 

Can be used in following manners:
- ScatterPlots
- HexbonPlots
- Kdeplots
- Regplots


```python
tips.head()
```

```python
#joint plot on total_bill and tip
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='hex',data=tips) # a 2d histogram- hexplot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

Distplots
- used to univariate analysis on continuous/numerical data
- histogram, PDF, Rugplot

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
    sns.distplot(titanic['age'])
```

```python
#plotiing only histogram
sns.distplot(titanic['age'],kde=False)
```

```python
#plotiing Only PDF
sns.distplot(titanic['age'],kind='kde')
```

```python
#plotting only histogram
sns.displot(titanic['age'],bins=100)
```

```python
sns.distplot(titanic['age'],rug=True,kde=False,hist=False)
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
